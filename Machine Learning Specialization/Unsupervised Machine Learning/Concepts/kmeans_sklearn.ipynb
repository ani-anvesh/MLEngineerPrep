```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import numpy as np

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Step 1: Run K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X)

# Step 2: Run PCA to reduce dimensions to 2D for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Step 3: Plot the clusters along with true labels
plt.figure(figsize=(12, 5))

# Plot predicted clusters
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', edgecolor='k', s=50)
plt.title('K-Means Clusters on Iris (PCA-reduced)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')

# Plot true labels for comparison
plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.title('True Iris Labels (PCA-reduced)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')

plt.show()
```

```python
from sklearn.datasets import load_digits

# Load Digits dataset
digits = load_digits()
X_digits = digits.data
y_digits = digits.target

# Step 1: Run K-Means clustering (10 clusters for digits 0-9)
kmeans_digits = KMeans(n_clusters=10, random_state=42)
clusters_digits = kmeans_digits.fit_predict(X_digits)

# Step 2: PCA to 2 components for visualization
pca_digits = PCA(n_components=2)
X_digits_pca = pca_digits.fit_transform(X_digits)

# Step 3: Plot the clusters and true labels
plt.figure(figsize=(12, 5))

# Predicted clusters
plt.subplot(1, 2, 1)
plt.scatter(X_digits_pca[:, 0], X_digits_pca[:, 1], c=clusters_digits, cmap='tab10', edgecolor='k', s=50)
plt.title('K-Means Clusters on Digits (PCA-reduced)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')

# True labels
plt.subplot(1, 2, 2)
plt.scatter(X_digits_pca[:, 0], X_digits_pca[:, 1], c=y_digits, cmap='tab10', edgecolor='k', s=50)
plt.title('True Digit Labels (PCA-reduced)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')

plt.show()
```
